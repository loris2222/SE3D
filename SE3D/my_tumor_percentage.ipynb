```python
from pathlib import Path 
import numpy as np

base_path= Path("/home/lorisg96/SE3D/")

thresholds = [-1, 0, 0.001, 0.1, 0.2, 0.3, 0.5, 0.8, 1, 1.2, 1.5, 1.8, 2]
flag = 0

for thresh in thresholds:
    all_accuracy = []
    tumor_accuracy = []
    false_negatives = []

    all_supports = []
    tumor_supports = []

    for i in range(5):
        cc_path = base_path / (f"correctly_classified{i}.npy")
        pr_path = base_path / (f"percentages{i}.npy")
        
        corr_classified = np.load(cc_path)
        percentages = np.load(pr_path)
        
        all_acc_fold = np.sum(corr_classified) / corr_classified.size
        tumor_corr_classified = corr_classified[percentages>thresh]
        tum_acc_fold = np.sum(tumor_corr_classified)
        fn_fold = np.sum(corr_classified[percentages==0]==0)

        all_accuracy.append(all_acc_fold)
        tumor_accuracy.append(tum_acc_fold)
        false_negatives.append(fn_fold)
        all_supports.append(corr_classified.size)
        tumor_supports.append(tumor_corr_classified.size)

    if not flag:
        print(f"all supports: {all_supports}")
        print(f"Global accuracy: {np.mean(all_accuracy)} +- {np.std(all_accuracy)}")
        print(f"fn: {np.sum(false_negatives) / np.sum(all_supports)}")
        flag=1

    print(f"Thresh: {thresh}")
    print(f"Supports:  {tumor_supports}")
    print(f"Accuracy over thresh: {np.sum(tumor_accuracy) / np.sum(tumor_supports)}")
```
